### Merging, Cleaning, & Dropping Data

#### Drop Extra Columns (anything with '_RANK')

```python
nba_trad_df = nba_trad_df[nba_trad_df.columns.drop(list(nba_trad_df.filter(regex='_RANK')))]
nba_adv_df = nba_adv_df[nba_adv_df.columns.drop(list(nba_adv_df.filter(regex='_RANK')))]
```

#### Merge DataFrames

The nba.stats.com serves advanced and traditional stats from different endpoints, so they will have to be merged.  Rows will be combined on **GAME_ID** and **TEAM_ABBREVIATION** since each game will have a unique identifier, and each game has two rows of data (one for each team).

The preliminary list of matching columns is:
    - SEASON_YEAR
    - MIN
    - TEAM_ID
    - TEAM_ABBREVIATION
    - TEAM_NAME
    - GAME_ID
    - GAME_DATE
    - MATCHUP
    - WL

```python
dup_columns = ["SEASON_YEAR", "MIN", "TEAM_ID", "TEAM_ABBREVIATION", 
               "TEAM_NAME", "GAME_ID", "GAME_DATE", "MATCHUP", "WL"]
merged = nba_trad_df.merge(nba_adv_df, how='inner', on=dup_columns)
```

```python
merged.head(15)
```

#### Drop Columns

We don't need all of the columns (FGM, FGA, etc), and some of the columns directly relate to outcome (PTS). Additionally, the outcome of the game is hidden in OFF_RATING, DEF_RATING, NET_RATING, and a couple others, so those will have to be dropped as well.

```python
columns_to_drop = ['SEASON_YEAR', 'TEAM_ID', 'TEAM_NAME', 'GAME_DATE', 'MIN', 'FGM',
                   'FGA', 'FG3M', 'FG3A', 'FTM', 'FTA', 'BLKA', 'PFD', 'PTS', 'AST_TO',
                   'E_OFF_RATING', 'OFF_RATING', 'E_DEF_RATING', 'DEF_RATING', 
                   'AST_RATIO', 'E_NET_RATING', 'NET_RATING', 'E_PACE', 'PACE', 'PIE']
nba_df = merged.drop(columns_to_drop, axis=1)
```

```python
for col in nba_df.columns:
    print(col)
```

1. Add "Home" field to keep track of the home team
2. Access "Games" (via GAME_ID) as pairs of rows
3. Create "NET_[stat]" columns that are the NET for home team (home team rebounds - away team rebounds)
4. Drop "Away" rows so that we have one row per game

```python
# positions home/away rows for each game together
nba_df.sort_values(['GAME_ID'], inplace=True)
```

#### Add Home Team Attribute

Since each game has two rows, one for the away and home team, I'm going to add a "HOME" attribute for the home teams, and later, I'll be dropping the away team, so that I am only looking at home games.

```python
# 1 = HOME, 0 = AWAY
nba_df['HOME'] = pd.np.where(nba_df.MATCHUP.str.contains("vs."), 1, 0)
```

```python
nba_df.head()
```

#### Set Feature Columns (to calculate differences)

```python
# cols to find differences (ex: Home REBs - Away REBs)
cols_to_diff = ["FG_PCT", "FG3_PCT", "FT_PCT", "OREB", "DREB", 
                "REB", "AST", "TOV", "STL", "BLK", "PF", 
                "AST_PCT", "OREB_PCT", "DREB_PCT", "REB_PCT", 
                "TM_TOV_PCT", "EFG_PCT", "TS_PCT"
               ]
```

Iterate through DataFrame ordered by **GAME_ID**, find home team, calculate difference between home team's stats and away team's stats and add column "NET_[

```python
data = nba_df.copy()
for i in range(0, len(nba_df), 2):
    gid = data.iloc[i]['GAME_ID']
    for col in cols_to_diff:
        home_stat = data.loc[(data['GAME_ID'] == gid) & (data['HOME'] == 1), col].values[0]
        away_stat = data.loc[(data['GAME_ID'] == gid) & (data['HOME'] == 0), col].values[0]
        stat_diff = home_stat - away_stat
        new_col = "NET_" + col
        data.loc[(data['GAME_ID'] == gid) & (data['HOME'] == 1), new_col] = stat_diff
```

```python
data.head(10)
```

### Drop Non-"NET"  Columns & Extra Columns

```python
# drop the cols used to get "NET_[col]"
data.drop(cols_to_diff, axis=1, inplace=True)
```

```python
# drop away team rows which now have NaNs
data.dropna(inplace=True)
```

```python
# the extra columns
data.drop(['TEAM_ABBREVIATION', 'GAME_ID', 'HOME'], axis=1, inplace=True)
```

## EDA

### Create Subsets for Visualization

```python
eda_df = data.drop(['MATCHUP', 'NET_OREB_PCT', 'NET_DREB_PCT', 'NET_AST_PCT', 'NET_TM_TOV_PCT'], axis=1)

eda_df_w = eda_df.loc[eda_df["WL"] == "W"]
eda_df_l = eda_df.loc[eda_df["WL"] == "L"]
print("Total Games: ", len(eda_df))
print("Total (Home) Wins: ", len(eda_df_w))
print("Total (Home) Losses: ", len(eda_df_l))
print("% (Home) Games Won: ", (len(eda_df_w)/len(eda_df)))
```

### Distribution Plot of Team Statistics

```python
feature_names = ["NET_FG_PCT", "NET_FG3_PCT", "NET_FT_PCT", 
                 "NET_OREB", "NET_DREB", "NET_REB", "NET_AST",
                "NET_TOV", "NET_STL", "NET_BLK", "NET_REB_PCT",
                "NET_EFG_PCT", "NET_TS_PCT", "NET_PF"]

@interact
def distribution_preview(feature = feature_names):
    sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [14, 5]})
    fig1 = plt.figure(figsize=(14,3))
    p = sns.distplot(eda_df[feature], label=feature, bins=50)
    plt.title("Distribution of {} (All Game Outcomes)".format(feature))
    p.legend()
    
    fig2 = plt.figure(figsize=(14,3))
    p = sns.distplot(eda_df_w[feature], label="{} in W".format(feature), bins=50)
    sns.distplot(eda_df_l[feature], label="{} in L".format(feature), bins=50)
    plt.title("Distribution of {} (W vs. L)".format(feature))
    p.legend()
    
    plt.show()
```

### Swarm Plot

```python
def quantitative_summarized(dataframe, x=None, y=None, hue=None, palette='Set1', ax=None, verbose=True, swarm=False):
    '''
    Helper function that gives a quick summary of quantattive data
    Arguments
    =========
    dataframe: pandas dataframe
    x: str. horizontal axis to plot the labels of categorical data (usually the target variable)
    y: str. vertical axis to plot the quantitative data
    hue: str. if you want to compare it another categorical variable (usually the target variable if x is another variable)
    palette: array-like. Colour of the plot
    swarm: if swarm is set to True, a swarm plot would be overlayed
    Returns
    =======
    Quick Stats of the data and also the box plot of the distribution
    '''
    series = dataframe[y]
    print(series.describe())
    print('mode: ', series.mode())
    if verbose:
        print('='*80)
        print(series.value_counts())

    sns.boxplot(x=x, y=y, hue=hue, data=dataframe, palette=palette, ax=ax)

    if swarm:
        sns.swarmplot(x=x, y=y, hue=hue, data=dataframe,
                      palette=palette, ax=ax)

    plt.show()
```

```python
feature_names = ["NET_FG_PCT", "NET_FG3_PCT", "NET_FT_PCT", 
                 "NET_OREB", "NET_DREB", "NET_REB", "NET_AST",
                "NET_TOV", "NET_STL", "NET_BLK", "NET_REB_PCT",
                "NET_EFG_PCT", "NET_TS_PCT", "NET_PF"]

@interact
def swarm_preview(feature = feature_names):
    sns.set(style='whitegrid', palette="deep", font_scale=1.1, rc={"figure.figsize": [16, 6]})
    quantitative_summarized(dataframe= eda_df, y = 'WL', x = 'NET_FG_PCT', palette="Set2", verbose=False, swarm=True)
```
