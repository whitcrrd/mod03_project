### Load Traditional & Advances Stats (JSON)

```python
nba_data_trad = json.load(open('data/2018-2019-team-game-logs-traditional.json'))
nba_data_adv = json.load(open('data/2018-2019-team-game-logs-advanced.json'))
```

#### Extract Headers/Columns

```python
columns_trad = nba_data_trad['resultSets'][0]['headers']
columns_adv = nba_data_adv['resultSets'][0]['headers']
print("TRADITIONAL: ", columns_trad[:5])
print("ADVANCED: ", columns_adv[:5])
```

#### Extract Rows

```python
rows_trad = nba_data_trad['resultSets'][0]['rowSet']
rows_adv = nba_data_adv['resultSets'][0]['rowSet']
rows_trad[:5]
```

#### Create Initial DataFrames

```python
nba_trad_df = pd.DataFrame(rows_trad, columns=columns_trad)
nba_adv_df = pd.DataFrame(rows_adv, columns=columns_adv)
```

```python
nba_trad_df = nba_trad_df[nba_trad_df.columns.drop(list(nba_trad_df.filter(regex='_RANK')))]
```

```python
for col in nba_trad_df.columns:
    print(col)
```

```python
nba_adv_df = nba_adv_df[nba_adv_df.columns.drop(list(nba_adv_df.filter(regex='_RANK')))]
```

```python
for col in nba_trad_df.columns:
    print(col)
```

#### Drop Extra Columns (anything with '_RANK')

```python
nba_trad_df = nba_trad_df[nba_trad_df.columns.drop(list(nba_trad_df.filter(regex='_RANK')))]
nba_adv_df = nba_adv_df[nba_adv_df.columns.drop(list(nba_adv_df.filter(regex='_RANK')))]
```

#### Merge DataFrames

The nba.stats.com serves advanced and traditional stats from different endpoints, so they will have to be merged.  Rows will be combined on **GAME_ID** and **TEAM_ABBREVIATION** since each game will have a unique identifier, and each game has two rows of data (one for each team).

The preliminary list of matching columns is:
    - SEASON_YEAR
    - MIN
    - TEAM_ID
    - TEAM_ABBREVIATION
    - TEAM_NAME
    - GAME_ID
    - GAME_DATE
    - MATCHUP
    - WL

```python
dup_columns = ["SEASON_YEAR", "MIN", "TEAM_ID", "TEAM_ABBREVIATION", "TEAM_NAME", "GAME_ID", "GAME_DATE", "MATCHUP", "WL"]
merged = nba_trad_df.merge(nba_adv_df, how='inner', on=dup_columns)
# merged = nba_trad_df.merge(nba_adv_df, how='inner', on=['GAME_ID', 'TEAM_ABBREVIATION', 'TEAM_NAME', 'GAME_DATE', 'MATCHUP'])
```
