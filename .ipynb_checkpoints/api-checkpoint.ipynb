### Load Traditional & Advances Stats (JSON)

```python
nba_data_traditional = json.load(open('data/2018-2019-team-game-logs-traditional.json'))
nba_data_advanced = json.load(open('data/2018-2019-team-game-logs-advanced.json'))
```

#### Extract Headers/Columns

```python
columns_traditional = nba_data_traditional['resultSets'][0]['headers']
columns_advanced = nba_data_advanced['resultSets'][0]['headers']
print("TRADITIONAL: ", columns_traditional[:5])
print("ADVANCED: ", columns_advanced[:5])
```

#### Extract Rows

```python
rows_traditional = nba_data_traditional['resultSets'][0]['rowSet']
rows_advanced = nba_data_advanced['resultSets'][0]['rowSet']
rows_traditional[:5]
```

#### Create Initial DataFrames

```python
nba_traditional_df = pd.DataFrame(rows_traditional, columns=columns_traditional)
nba_advanced_df = pd.DataFrame(rows_advanced, columns=columns_advanced)
```

#### Merge DataFrames

The nba.stats.com serves advanced and traditional stats from different endpoints, so they will have to be merged.  Rows will be combined on **GAME_ID** and **TEAM_ABBREVIATION** since each game will have a unique identifier, and each game has two rows of data (one for each team).
